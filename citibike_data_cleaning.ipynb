```python
#import dependencies
import pandas as pd
```

```python
# read csv
bike0319 = pd.read_csv("201903-citibike-tripdata.csv")
bike0320 = pd.read_csv("202003-citibike-tripdata.csv")
bike0319.head()
```

```python
bike0320.head()
```

```python
#Variables
#----------
# Trip Duration (seconds)
# Start Time and Date
# Stop Time and Date
# Start Station Name
# End Station Name
# Station ID
# Station Lat/Long
# Bike ID
# User Type (Customer = 24-hour pass or 3-day pass user; Subscriber = Annual Member)
# Gender (Zero=unknown; 1=male; 2=female)
# Year of Birth
```

```python
# inspect data
bike0319.describe() #1.327960e+06
bike0320.describe() #1.068457e+06
#combined data should result in 2396417 rows
```

```python
# calculate age
bike0319['age']=2019-bike0319['birth year']
bike0319.head()
bike0320['age']=2020-bike0320['birth year']
bike0320.head()
```

```python
bike0319["gender"].replace({0: "unknown", 1: "male", 2:"female"}, inplace=True)
bike0319.head()
bike0319['gender'].value_counts()
```

```python
bike0320["gender"].replace({0: "unknown", 1: "male", 2:"female"}, inplace=True)
bike0320.head()
bike0320['gender'].value_counts()
```

```python
bike0319["tripduration_min"] = round(bike0319["tripduration"]/60,2)
bike0319.head()
```

```python
bike0320["tripduration_min"] = round(bike0320["tripduration"]/60,2)
bike0320.head()
```

```python
# combine dataset
combined = pd.concat([bike0319, bike0320])
combined.head()
combined.describe() #comfirmed that data concatenated 
```

```python
bike0319['ride_id'] = bike0319.index
bike0319.head()

bike0320['ride_id'] = bike0320.index
bike0320.head()

combined['ride_id'] = combined.index
combined.head()
```

```python
bike0319.to_csv("citibike0319.csv")
bike0320.to_csv("citibike0320.csv")
combined.to_csv("combined_citibike.csv")
```
